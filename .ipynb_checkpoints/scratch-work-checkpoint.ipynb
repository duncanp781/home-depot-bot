```python
links = soup.find_all('a', href=True)
cat_links = [link['href'] for link in links if re.match(r'^/b/.*', link['href'])]
page_links = [link['href'] for link in links if re.match(r'^/p/.*', link['href'])]
#A search can give different categories, rather than a list of items.
#It seems like /b/ are category searches, and /p/ are specific pages
page_links
```

```python
#Takes a plain-text query, like 'grills'.
#Query-type can be a few different things. 'b' for categories, 'p' for pages
def get_links(query, query_type = 'b'):
    url = f'https://homedepot.com/s/{query}?NCNI-5'

    headers = { 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0' }
    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.text)
    links = soup.find_all('a', href=True)
    
    #The links in the anchors are relative, so start with /
    query_regex = r'^/' + re.escape(query_type)+ r'/.*'
    cat_links = [f'https://homedepot.com{link["href"]}' for link in links if re.match(query_regex, link['href'])]
    return cat_links
```

As we can see, there are way more pages if we search propane grill vs if we search for grills, since searching grills will prompt us with categories, but propane grills gives products.

```python
#This is a way to split apart the links for product pages
pages_split = [[link] + link.split('/')[3:] for link in prop_pages]
pages_split_dict = [{'link': page[0], 'page-type': page[1], 'name': ' '.join(page[2].split('-')[:-2])} for page in pages_split]
pages_split_dict
```

```python
@tool
def get_homedepot_pages(query):
    """Use this to search the homedepot website for a given query. 
    You should ask for specific terms.
    For example, 'propane grill' gives better results than 'grill'
    """
    page_links = get_links(query, 'p')
    pages_split = [[link] + link.split('/')[3:] for link in page_links[0]]
    def clean_name(name_part):
        return ' '.join(name_part.split('-'))
    pages_split_dict = [{'link': page[0], 'page-type': page[1], 'name': clean_name(page[2])} for page in pages_split]
    pages_strings = [ f'A product with name {page["name"]} is available at the url {page["link"]}. ' for page in pages_split_dict]
    return ''.join(pages_strings)
```

```python
tools = [get_homedepot_pages]
agent = initialize_agent(
    tools, 
    llm, 
    agent="chat-zero-shot-react-description",
    verbose=True,
    handle_parsing_errors=True
)
```

```python
agent.run("What is a url for a propane grill on the home depot website?")
```
